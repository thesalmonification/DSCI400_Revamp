**Imports**

```
import numpy as np
import pandas as pd
from scipy.fftpack import dct
import matplotlib.pyplot as plt
%matplotlib inline
```

**Plot the Discrete Cosine Transforms for Each EEG**

```
def plot_dcts(lower = 2, upper = 30):
  """
  This function loads the EEG data and plots the DCT of each EEG from the range
  (lower, upper - 1) based on Python indexing.
  """
  for i in range(2,12,2):
    try:
     df = pd.read_hdf('/content/drive/Shared drives/DSCI400_Revamp/Waveform_Vocoded_Equalized_Data.h5', 't' + str(i)).transpose()
    except:
      continue
    num_rows = df.count()[0]
    vals = np.array(df.index)
    for j in range(num_rows):
      vector = df.iloc[j,:].to_numpy()
      transform = dct(vector)
      plt.scatter(range(lower, upper), transform[lower:upper])
      plt.xlabel('Cosine frequency (Hz)')
      plt.ylabel('Value/Amount in Signal')
      plt.title('Session ' + str(i) + '; Channel ' + str(vals[j]))
      plt.show()
```

```
plot_dcts()
```
