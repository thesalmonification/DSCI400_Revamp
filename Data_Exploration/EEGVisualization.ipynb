Choose a session number to look at, by entering a number as a string (e.g. '2') (Hint: all sessions are pos.,even numbers between 2 and ~4000)

```python
#Choose a session to look at
session = '2'
```

```python
%matplotlib inline
#Grab the BDF file in the directory
bdf_files = glob.glob('/content/drive/Shared drives/DSCI400_Revamp/Sessions/'+session+'/*.bdf')

#Open the data using mne
try:
  raw = mne.io.read_raw_bdf(bdf_files[0], preload=True)
except:
  print('Sorry, you tried a session that did not have a BDF file')
#Display stats of EEG recording
print(raw.info)

#Display channel names
print(raw.ch_names) #Only the first 32 channels matter (others are not EEG)

#Visualize the EEG waveforms
raw.plot()
```
